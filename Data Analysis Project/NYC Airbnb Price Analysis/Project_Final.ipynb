```python
import statsmodels.api as sm


df['const'] = 1


reg1 = sm.OLS(endog=df['price'], exog=df[['const', 'number_of_reviews']], \
    missing='drop')
type(reg1)

results = reg1.fit()
type(results)

print(results.summary())
```

### The above is a OLS Regression Results between price and number_of_reviews. As we want to maximize your adjusted R squred but minimize AIC and BIC.
